#Question: You have been provided with a dataset called "Golf_Gaming.csv" which has information on whether golf will be played at the Namulonge Golf Course depending on various weather parameters

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib as mat
import matplotlib.pyplot as plt
import plotly.express as px
import warnings
warnings.filterwarnings("ignore")
```

READING CSV FILE

```python
golfing_df = pd.read_csv('C:/Users/katuk/Downloads/Golf_Gaming.csv')
golfing_df
```

REMOVING THE FIRST CHARACTER IN THE DAY COLUMN

```python
golfing_df['Day'] = golfing_df['Day'].str[1:]
golfing_df
```

CONVERTING THE WIND COLUMN TO NUMERIC VALUES USING PANDAS

```python
golfing_df['Wind'].replace(['Strong','Weak'],
                           [0,1], inplace=True)
golfing_df
```

CONVERTING THE HUMIDITY COLUMN TO NUMERIC VALUES USING PANDAS

```python
golfing_df['Humidity'].replace(['Low','Normal','High'],
                           [0,1,2], inplace=True)
golfing_df
```

CONVERTING THE TEMPERATURE COLUMN TO NUMERIC VALUES USING PANDAS

```python
golfing_df['Temperature'].replace(['Hot','Mild','Cool'],
                           [0,1,2], inplace=True)
golfing_df
```

CONVERTING THE OUTLOOK COLUMN TO NUMERIC VALUES USING PANDAS

```python
golfing_df['Outlook'].replace(['Sunny','Overcast','Rain'],
                           [0,1,2], inplace=True)
golfing_df
```

DROPPING LAST ROW CAUSE VALUES ARE NOT NECESSARY AT THIS POINT

```python
golfing_df = golfing_df.drop([16])
golfing_df
```

SPLITTING DATA TO PREDICT OUTPUTS

```python
X = golfing_df.drop(columns=["Play Golf"])
X
```

```python
y = golfing_df['Play Golf']
y
```

BUILDING THE MODEL USING THE DECISION TREE METHOD

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

TRAINING THE MODEL USING THE INPUT SET(X) AND OUTPUT SET(y)

```python
model.fit(X, y)
```

MAKING PREDICTIONS

```python
prediction = model.predict([[17, 1, 0, 0, 0]])
prediction
```

TESTING THE MODEL

```python
# test model for 20% with day
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

prediction = model.predict(X_test)
 
score = accuracy_score(y_test, prediction)
score
```
